```python
#Create Visual for the top five most popular beverage types for each year based on retail sales
# Set up the matplotlib figure
def top5byyear(year, data_set):
    data_df = data_set.loc[data_set['year']==year].sort_values(by='count', ascending=False).head()
    x = data_df['type_id'].astype(str)
    y = data_df['count']
    sns.set_theme(style="white", context="talk")
    # f, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(7, 5), sharex=True)
    
    # access the sequential data
        # x = np.array(list("ABCDEFGHIJ"))
        # y1 = np.arange(1, 11)
    sns.barplot(x=x, y=y, hue=x, palette="rocket").set_title(year)
    #ax1.axhline(0, color="k", clip_on=False)
    #ax1.set_xlabel()#use variable containing the year the data was pulled from
    #ax1.set_ylabel("Retail Sales (By Case)")
    
    # Finalize the plot
    sns.despine(bottom=True)
    # plt.setp(f.axes, yticks=[])
    # plt.tight_layout(h_pad=2)
    #export graph
    plt.savefig(f'top5_types_{year}.png')
```

```python
top5byyear(2020, type_counts)
```
